```
courses = []
choice = 'random'

def line():
    print('------------------')

def menu():
    line()
    print('MENU')
    line()
    print('1. Add course')
    print('2. Remove course')
    print('3. View courses')
    print('4. Exit')
    line()
    choice = input('Enter your choice: ')
    return choice

while choice != '4':
    choice = menu()
    if choice == '1':
        course = input('Enter course name :  ')
        courses.append(course)
        print('Course', course, 'added\n')
    elif choice == '2':
        course = input('Enter removed crouse name :D ')
        if course in courses:
            courses.remove(course)
            print(course,'course removed' )
        else:
            print('Not found', course)
    elif choice == '3':
        print('View the courses:')
        for course in courses:
            print(course)
    elif choice == '4':
        print('Thank you\n')
    else:
        print('Please enter one of 1 to 4 \n')
```
